### Plot the distributions of two classes, 1. GDP < 58000 and  2. GDP >= 58000, of COVID-19 State Data using different colors and different  markers where x-axis is the Pollution and y-axis the Mortality-rate.

```python
import matplotlib.pyplot as plt
```

```python
calculate_mortality_rate = lambda row: row['Deaths']/row['Population']*100
```

```python
mortality_rate_col = data.apply(calculate_mortality_rate, axis=1)
```

```python
data = data.assign(**{'Mortality-rate': mortality_rate_col.values})
```

```python
subData1 = data[data['GDP']<58000]
subData2 = data[data['GDP']>=58000]
```

```python
plt.xlabel('Pollution')
plt.ylabel('Mortality Rate')
plt.plot(subData1['Pollution'], subData1['Mortality-rate'], 'go')
plt.plot(subData2['Pollution'], subData2['Mortality-rate'], 'rx')
plt.legend(['GDP < 58000', 'GDP >= 58000'])
plt.show()
```

# Step 7
### Show the proportion of following three classes of COVID-19 State Data using pie chart.

**<font color = blue size=3> About the class: </font>** <p>
 1、 Mortality-rate < 0.02 <p>
 2、 Mortality-rate between 0.02 and 0.03 <p>
 3、Mortality-rate > 0.03 <p>

```python
labels = ["MR <= 0.02", "0.02 < MR < 0.03", "MR >= 0.03"]
```

```python
mr = [data[data['Mortality-rate']<=0.02].shape[0], data[(data['Mortality-rate']>0.02) & (data['Mortality-rate']<0.03)].shape[0], data[data['Mortality-rate']>=0.03].shape[0]]
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(mr, labels = labels, autopct='%1.1f%%')
ax1.axis('equal')
fig1.show
```
